# Introduction

Customer segmentation (or customer personality analysis) is the process of dividing customers into groups on common characteristics, such as demographics characteristics or buying behaviors.

Customer personality analysis helps a business to modify its product based on its target customers from different types of customer segments. For example, instead of spending money to market a new product to every customer in the company’s database, a company can analyze which customer segment is most likely to buy the product and then market the product only on that particular segment.

In this notebook, we analyze a dataset from an anonymized company in the US that has collected many such features from its customers, such as age, education, marital status, income, amount spent on their stores and whether or not they have responded positively to marketing campaigns and deals. 

The goal of this project is to use Clustering Machine Learning Techniques in order to find the best customer segmentation, and then find and analyze what characteristics those clusters have in common.

# Data Analysis and Feature Engineering

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('marketing_campaign.csv', sep='\t')
df.head()
```

Let's take a look at the dataframe's columns and the attribute information. 

```python
df.info()
```

Attributes

People

- ID: Customer's unique identifier
- Year_Birth: Customer's birth year
- Education: Customer's education level
- Marital_Status: Customer's marital status
- Income: Customer's yearly household income
- Kidhome: Number of children in customer's household
- Teenhome: Number of teenagers in customer's household

Products

- MntWines: Amount spent on wine in last 2 years

Promotion

- NumDealsPurchases: Number of purchases made with a discount

- AcceptedCmp3: 1 if customer accepted the offer in the 3rd campaign, 0 otherwise
- AcceptedCmp4: 1 if customer accepted the offer in the 4th campaign, 0 otherwise
- AcceptedCmp5: 1 if customer accepted the offer in the 5th campaign, 0 otherwise

- NumWebPurchases: Number of purchases made through the company’s website
- NumCatalogPurchases: Number of purchases made using a catalogue
- NumStorePurchases: Number of purchases made directly in stores

Let's first review some features in order to make sure that the data is correct.

There are two features whose information was not provided, 'Z_CostContact' and 'Z_Revenue'. We can check their values as follows.

```python
df['Z_CostContact'].unique()
```

```python
df['Z_Revenue'].unique()
```

It seems that these two features are completely irrelevant, so we can just drop them. And we will also drop the 'ID' column since we don't need it.

```python
df.drop(['ID', 'Z_CostContact', 'Z_Revenue'], axis=1, inplace=True)
```

As we saw with df.info() before, there are some missing values in the 'Income' feature. We will drop these rows. After that, we should have no missing values left.

```python
df.dropna(inplace=True)
print('Sum of missing values per column:')
df.isnull().sum()
```

Now, since models can't be trained on datetime data, we will create a new feature from Dt_Customer that will simply count the days a customer has been enrolled in the company. For simplicity, we will also assume that for the newest customer, 0 days have passed since their enrollment. 

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
dates = []
days = []


for customer_date in df["Dt_Customer"]:
    # date customers enrollment
    dates.append(customer_date.date())
    
    
# newest customer
most_recent_date = max(dates)

    
for customer_date in df["Dt_Customer"]:
    # days since customer enrollment 
    # 0 days for the newest customer
    days.append(most_recent_date - customer_date.date())
```

```python
df['Customer Days'] = days
df['Customer Days'] = df['Customer Days'].dt.days
```

```python
df[['Dt_Customer', 'Customer Days']].head()
```

After that, we can drop the 'Dt_Customer' column.

```python
df.drop('Dt_Customer', axis=1, inplace=True)
```

Next, we will replace the feature 'Year Birth' with the customer's age. For simplicity, we will assume that we're in 2014 and everyone's birthday has already passed this year.

```python
# Age of customer in 2014
df['Age'] = 2014-df['Year_Birth']
df[['Year_Birth', 'Age']].head()
```

Then we can drop the 'Year_Birth' column.

```python
df.drop('Year_Birth', axis=1, inplace=True)
```

Let's now pay attention to the categorical features 'Marital_Status' and 'Education'.

```python
df['Marital_Status'].value_counts()
```

To simplify things, we'll lump 'Widow', 'Divorced' and the other ones together with 'Single'.

```python
df["Marital_Status"].replace({'Alone':'Single', 
                              'Absurd':'Single', 
                              'YOLO':'Single',
                              'Widow':'Single',
                              'Divorced':'Single'}, inplace=True)
```

```python
df['Marital_Status'].value_counts()
```

Let's explore now 'Education'.

```python
df['Education'].value_counts()
```

As we did with 'Marital_Status', let's lump together some of those in order to simply have three values.

```python
df["Education"].replace({"Basic":"Undergraduate",
                                         "2n Cycle":"Undergraduate", 
                                         "Graduation":"Graduate", 
                                         "Master":"Postgraduate", 
                                         "PhD":"Postgraduate"}, inplace=True)
```

```python
df['Education'].value_counts()
```

Now, to visualize better how every kind of customer spends their money, we'll create a new feature to reflect the total amount they have spent during the last 2 years.

```python
products_spent = ['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']

df['Total_Spent'] = np.sum(df[products_spent], axis=1)
```

```python
df[products_spent+['Total_Spent']].head()
```

Before we compare the new feature 'Total_Spent' with other features like 'Age' and 'Income', let's see if there are some outliers in the latter ones. 

```python
fig, axes = plt.subplots(1, 3, dpi=200, figsize=(9,4))

sns.boxplot(ax=axes[0], data=df, y='Age')
sns.boxplot(ax=axes[1], data=df, y='Income', color='darkblue')
sns.boxplot(ax=axes[2], data=df, y='Total_Spent', color='lightblue')

fig.subplots_adjust(wspace=.6);
```

Unexpectedly, we've got several people who are older than 100. We will drop these outliers, as well as rows with Income higher than 600,000 dollars.

```python
df = df[(df['Age']<100) & (df['Income']<600000)]
```

Let's now make a pairplot to compare 'Age', 'Income' and 'Total_Spent'.

```python
sns.pairplot(df[['Income','Age', "Total_Spent"]]);
```

From here, we can infer that there doesn't seem to be a significant correlation between 'Age' and 'Total_Spent', nor 'Age' and 'Income'. However, 'Income' and 'Total_Spent' seem to be positively correlated. Let's make a scatterplot between these two, but this time let's further get rid of more outliers.

```python
temp_df = df[df['Income']<100000]

sns.scatterplot(data=temp_df, x='Income', y='Total_Spent')
```

```python
# computing Pearson correlation
temp_df[['Income', 'Total_Spent']].corr()
```

As expected, there's a high correlation between 'Income' and 'Total_Spent'.

Now, we do the same thing but this time we will take into account the categorical features.

```python
fig, axes = plt.subplots(3, 1, dpi=100, figsize=(5,10))

sns.boxplot(ax=axes[0], data=df, x='Marital_Status', y='Age')
sns.boxplot(ax=axes[1], data=df, x='Marital_Status', y='Income')
sns.boxplot(ax=axes[2], data=df, x='Marital_Status', y='Total_Spent')

fig.subplots_adjust(hspace=.6);
```

From the above boxplots, we can infer that there are not significant differences between the three marital statuses.

Let's see what happens with 'Education'.

```python
fig, axes = plt.subplots(3, 1, dpi=100, figsize=(5,10))

sns.boxplot(ax=axes[0], data=df, x='Education', y='Age', palette='Set2')
sns.boxplot(ax=axes[1], data=df, x='Education', y='Income', palette='Set2')
sns.boxplot(ax=axes[2], data=df, x='Education', y='Total_Spent', palette='Set2')

fig.subplots_adjust(hspace=.6);
```

Here we can see that graduates and postgraduates tend to be a little older than undergraduates, and they also have higher income than undergraduates. Regarding their total spents, undergraduates also tend to spend less, with a few exceptions as outliers.

To finish the Exploratory Data Analysis, we'll look at the heatmap of the correlations. But since there are many features, we'll skip the features we used to compute 'Total_Spent'.

```python
plt.figure(figsize=(15,10))
correlations = round(df.drop(products_spent,axis=1).corr(), 2)
sns.heatmap(correlations, annot=True);
```

Here is where we see really interesting information:

- 'Income' is very negatively correlated with Kidhome (having 1 or 2 kids), but not with Teenhome (having 1 or 2 teenagers). Also, 'Income' is highly correlated with purchases, but not with having many visits on the company's website. This could mean that they spend less time browsing and more time actually buying stuff. 

- NumDealsPurchases (purchases with a discount) are slightly positively correlated with having kids and teenagers, as well we how often they visit the website. This would make sense as most people get to know about discounts through the website. 

- Customer Days (Days since the customer enrolled in their company) doesn't actually seem to matter much regarding the purchases they make. 

- Age is negatively correlated with having kids, but positively correlated with having teenagers, which makes sense.

- As expected, the Total Spent is positively correlated with Income, Purchases in general, but not with the amount of times they visit the website. But it's also correlated with some of the previous marketing campaigns. 

This concludes the EDA part.
